#연산자

##사칙연산자


더하기 : +  
빼기 : -  
곱하기 : *  
거듭제곱 : **  
나누기 : /  
나누기 몫 : //  
나누기 나머지 : %  


연산자 우선순위
 - 괄호 우선
 - 부호 > 곱,분 > 합,차

##비교연산자

동일하다 : ==  
동일하지 않다 : !=
왼쪽이 크다 : >  
왼쪽이 크거나 같다 : >=  
오른쪽이 크다 : <  
오른쪽이 크거나 같다 : <=

##논리 연산자

and : 둘 다 참일 때만 True

or : 둘 중 하나라도 참이면 True

not : 논리 반전

##대입 연산자(할당 연산자)
숫자형 자료의 대입 연산

= : 왼쪽 변수에 오른쪽 값을 할당  
+= : 왼쪽 변수에 오른쪽 값을 더하며 할당  
-= : 왼쪽 변수에 오른쪽 값을 빼며 할당  
*= : 왼쪽 변수에 오른쪽 값을 곱하며 할당  
/= : 왼쪽 변수에 오른쪽 값을 나누며 할당  

##멤버 연산자

a in [] : a가 []안에 포함되어 있는가  
a not in [] : a가 [] 안에 포함되어 있지않은가



##삼항 연산자

[True_value] if [condition] else [False_value]  
의미 : [condition]을 만족하면 [True_value]이고 아니면 [False_value]



##대화식 프로그래밍

```python
num = int(input('숫자를 입력하시오 : '))
'짝수입니다' if num%2==0 else '홀수입니다'
```

```python
num = int(input('숫자를 입력하시오 : '))
'숫자 {}은(는) {}'.format(num,'짝수입니다' if num%2==0 else '홀수입니다')
```

K씨의 성적은 국어 95점, 수학 50점, 영어 70점이다. K씨의 평균 점수를 구하라  
단, 국어 kor, 수학 math, 영어 eng 변수에 할당하고  
결과값은 score 변수에 할당하여 결과값을 출력하라.

```python
kor = 90; math = 50; eng = 70
score = (kor + math + eng)/3
print(score)
```

입력한 숫자가 10과 비교하여 큰 수를 출력하는 프로그램을 작성하라.
- num1 변수에는 10을 할당
- num2 변수에 대화형식으로 입력한 값을 할당
- 결과값은 '10과 {입력한 수} 중 큰 수는 {결과값}이다.' 가 나오도록 하시오.


```python
num1 = 10
num2 = int(input('숫자를 입력하시오 : '))
'{}과 {}중 큰 수는 {}이다'.format(num1,num2, '10' if 10>=num2 else num2)
```
