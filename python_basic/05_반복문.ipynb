#반복문

어떤 문장을 반복해서 수행해야할 때 while이나 for를 사용
- while : 조건식으로 반복수행
- for : 범위로 반복수행

##while문

- 조건식으로 하위 블럭 (실행문) 반복 실행
- 공백으로 들여쓰기에 의한 하위 블럭 구분

```
변수 초기값
while 조건식:
    실행할 문장
    변수 더하기

```

````python
num = 1
while num <= 4:
    print(num)
    num = num + 1
````

````python
i = 0 #변수 초기값
while i < 10: #변수가 범위 내에 있으면
    print('hello', i)
    i += 1 #할당연산자
````

###break

반복문에서 탈출 할 때 사용. 아래는 무한 반복문이다.


````python
i = 0 # 초기값 변수
while True: #무한 반복
    print('hello',i)
    i = i + 1 #변수에 1 더하기
````

````python
i = 0 # 초기값 변수
while True: #무한 반복
    if i >= 4 : #변수가 범위 내에 없으면 조건문에서 탈출한다.
        break
    print('hello',i)
    i = i + 1 #변수에 1 더하기
````

break 예문
- 자판기에 초콜렛이 10개 있다.
- 돈을 넣으면 초콜렛을 주고, 초콜렛의 갯수는 1개씩 차감한다.
- 이를 초콜렛의 갯수가 0개가 될때까지 반복수행한다.


````python
money = 1000
bool(money)
````

````python
 '='*40
````

````python
choco = 10
money = 1000
while money: # money는 bool타입이 True이므로 무한반복수행한다.
    print('자판기에 돈을 넣으면 초콜렛을 준다.')
    # choco -= 1
    choco = choco - 1
    print(f'남은 초콜렛의 양은 {choco}개이다.') #수정
    # print('남은 초콜렛의 양은 %d개이다.' % choco)
    print('='*40)
    if choco == 0: # choco가 =이면 중단한다.
        print('초콜렛이 품절되었다. 판매를 중지한다.')
        break
````

###continue


반복문의 처음으로 간다.

````python
num = 0
while num < 5:
    num = num + 1 # 수행문장
    if num == 3:
        continue # num이 3일 때 처음으로 돌아감
    else:
        print(num) # 3일 때는 출력이 안됨
````

##for문

- 범위로 하위 블럭 (실행문) 반복 실행
- 공백으로 들여쓰기에 의한 하위 블럭 구분

```
for 변수 in 리스트(문자열 등) :
    실행할 문장 1
    수행할 문장 2
```



````python
for num in range(1,5): # 조건에 만족하지 않을 때 중단
    print(num)
````

range() : range(start, stop, step)

````python
range(10)
````

````python
list(range(10))
````

````python
list(range(1,10,1)) #시작,끝,간격
````

````python
list(range(1,10,2))
````

변수의 범위 지정: 자료형으로 범위 주기



````python
for w in 'apple':
    print(w)
````

````python
for w in ['apple','banana']:
    print(w)
````

###이중사용


````python
for i in range(2,10): #2~9        # 첫번째 for문
    for j in range(1, 10):   # 두번째 for문
        # print(i*j)
        print(i*j, end=' ')
        # 매개변수 end를 사용한 이유는 출력결과문을 한줄에 쓰기 위함
    print('')
````

````python
print(1, end='')
print(2, end=' ')
print(3)
print(4)
````

````python
print('1')
print('')
print('2')
````

````python
print('1',end=' ')
print('2')
print('3')
````

###break


````python
for i in range(5): #0~4
    if i==3:
        break #해당 조건에서 반복문 중단.
    print(i, end = '')
````

###continue

````python
for i in range(5): # 0~4까지 반복
    if i == 3:
        continue # 다음 단계
    print(i,'번 반복')
````

###pass

````python
for i in range(5): # 0~4까지 반복
    if i==3:
        pass # 아래코드 수행
    print(i,'번 반복')
````

````python
# pass 활용예시
for i in range(5): # 0~4까지 반복
    if i==3:
        pass # 아래코드 수행
    else:
        print(i,'3이 아닙니다.')
    print(i,'번 반복')
````

##연습문제

문제) A 학급에 총 10명의 학생이 있다. 이 학생들의 중간고사 점수는 다음과 같다.
[70, 60, 55, 75, 95, 90, 80, 80, 85, 100]
- for문을 사용하여 A 학급의 평균 점수를 구해 보자.
- 단, 성적의 합은 for문으로 작성하고 for문 밖에서 평균성적을 구한다.
- 평균 = 합계 / 성적 개수, 성적 개수는 len()함수 사용


````python
A = [70, 60, 55, 75, 95, 90, 80, 80, 85, 100]
len(A)
a=0
for t in A:
    a += t
a/len(A)
````
