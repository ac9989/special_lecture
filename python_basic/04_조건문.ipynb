#조건문

조건을 판단하고 수행하는데 씀
- if문 : 조건이 참일 때 수행
- if,else문 : 조건이 참, 거짓일 때 각각 수행
- if,elif,else문 : 다양한 조건을 넣을 수 있음
- 중첩된 if문 : 하나의 if문 안에 다른 if문을 넣을 수 있음

##조건문


조건이 참이면 하위 문장을 실행

````python
65 > 90
````

````python
if 95 > 90: # 조건문
    print('95>90') # 수행할 문장
````

````python
if 65 > 90: # 조건문
    print('65>90') # 수행할 문장
````

들여쓰기를 하지 않으면 if문에 해당되지 않음


````python
if 65 > 90: # 조건문
print('65>90') # 수행할 문장
````

###자료의 참,거짓


````python
print(bool('Python')) # 문자열이 있으면 True
print(bool(''))      # 문자열이 없으면 False
print(bool(' '))     # 공백도 문자열 True
print(bool([1,2,3])) # list가 있으면 True
print(bool([]))      # list가 없으면 False
print(bool(1))       # 1 = True
print(bool(0))       # 0 = False
````

1) 문자형의 참, 거짓

````python
string = 'apple'
# string = ''
if string:
    print('조건이 참일 때 출력')
````

논리연산: and, or, not
- and:둘 다 True일 때 True
- or: 둘 중 하나가 True일 때 True
- not: 논리 반전


````python
string = '' #False
if not string: # 조건문은 참조건 입력. not을 입력하면 not True, False일때 문장수행
    print('조건이 참일 때 출력')
````

2) List의 참 거짓

````python
pocket = ['money', 'card']
if pocket:
    print('pocket에 무엇인가 들어가 있다.')
````

````python
pocket = [] #False
if not pocket:
    print('pocket은 비어있다.')
````

성적이 90이상이면 '합격입니다.', 그렇지 않으면 '불합격입니다.'를 출력하라.


````python
score = 60
if score >= 90:
    print('합격입니다.')
else:
    print('불합격입니다.')
````

##if - elif - else 문



```
if 조건문1:
    수행할 문장1
    수행할 문장2
elif 조건문2:
    수행할 문장A
    수행할 문장B
else:
    수행할 문장ㄱ
    수행할 문장ㄴ

```



````python
money = 10000
if money < 5000:
    print('빵을 먹는다.')
elif money < 15000:
    print('돈가스를 먹는다.')
else:
    print('치킨을 먹는다.')
````

pass : 조건에 만족해도 아무일이 일어나지 않음


````python
money = 3000
if money < 5000:
    pass # 해당 조건문에 수행할 문장 없음
elif money < 15000:
    print('돈가스를 먹는다.')
else:
    print('치킨을 먹는다.')
````

##if문 중첩 (if문 안에 if문)

````python
hungry = False
money = 1000
if hungry:
    if money < 5000:
        print('김밥을 먹는다.')
    else:
        print('김치볶음밥을 먹는다.')
else:
    if money > 3000:
         print('캔맥주를 산다')
    else :
        print('집에 가야지 뭐, 탄산수나 살까')
````

##연습문제

문제) 음수/양수 판별 음수 양수를 판별하는 조건문을 완성하여라.
- 입력 숫자가 0이면 '00은 0 입니다.'를 출력
- 입력 숫자가 양수이면 '00은 양수 입니다.'를 출력
- 입력 숫자가 음수이면 '00은 음수 입니다.'를 출력
- input(), int(), format 함수 활용

````python
n = int(input('숫자를 입력하세요'))
if n==0:
    print(f'{n}은 0입니다')
elif n>0:
    print(f'{n}은 양수 입니다')
else:
    print(f'{n}은 음수 입니다')
````
