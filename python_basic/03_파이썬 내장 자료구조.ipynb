##튜플

리스트와 정말 비슷하지만 단일항으로 존재할 수 없고, 한번 만들어지면 수정이 불가능하다!

```python
tp1 = ()
tp1
```

```python
tp2 = (1, 2, 3)
tp2
```

```python
tp4 = 1 # 튜플이 아님.
tp4
```

```python
type(tp4)
```

```python
tp4 = (1)
type(tp4)
```

```python
tp4 = (1,) # 값이 1개인 튜플 생성시 뒤에 콤마 필수
type(tp4)
```

```python
tp5 = ('a', 'b', ('ab', 'cd')) # 튜플이 튜플을 가질 수 있다.
tp5
```

튜플은 요소 삭제 및 추가가 불가능하다

```python
tp1 = (1, 2, 3, 'a', 'b')
del tp1[0]
```

```python
tp1 = (1, 2, 3, 'a', 'b')
tp1[0] = 'c'
```

더하기, 곱하기는 가능하다!

```python
tp1 = (1,'a')
tp2 = (4,3)
tp1 + tp2
```

```python
tp2 = (3, 4)
tp2 * 3
```

##집합

중복을 허용하지 않음  
순서가 없음 : 인덱싱으로 값을 얻을 수 없음  
교집합, 합집합, 차집합을 구할 때 용이


```python
set()
```

```python
set([1,2,3])
```

```python
{1,2,3}
```

중복허용하지 않음.  
순서가 없음.

```python
s1 = {1,2,3}
s1[0] # set은 순서가 없기 때문에 인덱싱할 수 없음
```

교집합

```python
s1 = set([1, 2, 3, 4])
s2 = set([1, 2, 5, 6])
```

```python
s1.intersection(s2) # 교집합
```

```python
s1 & s2 # 교집합
```

합집합

```python
 s1.union(s2)
```

```python
s1 | s2 #|: shift + \
```

차집합

```python
s1.difference(s2) # s1 중 s2와 다른 요소
```

```python
s1 - s2
```
