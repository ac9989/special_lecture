문제10)
입력된 성적정수에 따라 성적등급을 부여하여라.
- 80이상~100 : A
- 61이상~80미만 : B
- 31이상~60미만 : C
- 0~30미만 : D  
힌트: input(), int() 함수 활용  

```python
s = int(input('님 몇점? : '))
if s > 100:
    print('거짓말ㄴ')
elif s>80:
    print('A')
elif s>=61:
    print('B')
elif s>=31:
    print('C')
else:
    print('D')
```
