##return

함수의 결과 값을 함수 밖에서도 변수로 사용

````python
def add(a, b):  # a, b는 매개 변수
    return a + b # return을 쓰면 결과값을 함수 밖에서도 변수로 사용 가능하다.
````

````python
 add(3, 4)       # 3, 4는 인수
````

````python
 result = add(3,4)
 result
````

##*args: 입력값의 개수가 유동적일 때

입력값의 갯수가 유동적일 때: for문 혼용


````python
total = 0 # 초기값 설정
for i in [1,2,3]: #args = (1,2,3,10)   # 함수안에서 사용할때는 *없이 변수명
    total = total + i
````

````python
total
````

````python
def add_many(*args):  # 매개변수 앞에 *을 붙이면 입력값을 전부 모아서 튜플로 만들어준다.
    total = 0 # 초기값 설정
    for i in args: #args = (1,2,3,10)   # 함수안에서 사용할때는 *없이 변수명
        total = total + i #
    return total
````

args는 매개변수 arguments영단어의 약자이며 관례적으로 사용한다.


````python
 add_many(1,2,3)
````

for문 vs def 함수


for문 구구단

````python
for i in range(2,10):        # 첫번째 for문
    for j in range(1, 10):   # 두번째 for문
        print(i*j, end=" ")  # 매개변수 end를 사용한 이유는 출력결과문을 한줄에 쓰기 위함
    print('')
````

함수 구구단

````python
# i * j : 2*1~n*9
def function(*arg):
    for i in arg:
        for j in range(1, 10):   # 두번째 for문
            print(i*j, end=" ")  # 매개변수 end를 사용한 이유는 출력결과문을 한줄에 쓰기 위함
        print('')
````

````python
function(2,3,4,5,6,7,8,9)
# function(2,6)
````

##global: 함수 밖의 변수 사용

함수는 함수 안에서 정의된 변수 또는 매개변수만 사용 가능

````python
out=1
def test():
    out = out + 1
    return out
````

````python
test()
````

global은 매개변수의 인수 입력 없이 함수 안에서 함수밖의 변수를 직접 사용하겠다는 명령어

````python
out= 1
def test():# 매개변수를 입력하지 않음
    global out
     # global은 함수밖의 변수를 그대로 사용하므로
     # 매개변수를 입력할 필요가 없다.
    out = out + 1
    return out
````

- 프로그래밍을 할때 global명령어는 피하는 것이 좋다.
- 외부 변수에 종속적인 함수는 그다지 좋은 함수가 아니다
- 함수는 독립적으로 존재하는 것이 좋다.

##연습문제

문제) 주어진 자연수가 홀수가 맞는지 판별해 주는 함수(is_odd)를 작성해 보자.
- 홀수이면 True
- 짝수이면 False 의 결과값이 나오도록 한다.

````python
def odd(num):
    if num%2 == 0:
        return False
    else:
        return True
````

````python
a= int(input('숫자 써 봐 : '))
odd(a)
````
