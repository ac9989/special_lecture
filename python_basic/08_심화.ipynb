#예외처리(try, except)


- try : 기본 수행 문장들 입력
- except : 문제가 발생하면 수행할 문장을 입력

```
try:
    실행 문장 1
    시행 문장 2
except 발생오류 as 오류 메시지 변수:
    실행 문장 a
    실행 문장 b
```



````python
def div(a,b): #division:나누기
    return a/b
````

- 분모에 위치하는 매개변수 b에 0을 할당해주면 오류가 발생


````python
div(1,0) # 1/0
````

- 첫번째 방법: if문 활용  
bool(0)이 False인 것을 활용한다.


````python
def div(a,b):
    if b == 0:
        return '계산식 표현 불가'
    else:
        return a/b

div(1,0)
````

특정 오류에 대한 예외처리
- except 예약어에 발생오류와 오류메세지 변수 사용


````python
def div(a,b):
    try:
        return a/b
    except ZeroDivisionError: #오류메세를 그대로 가져오기
        return '계산식 표현 불가'
div(1,0)
````

오류메세지를 하나의 변수로 쓸 수도 있다!

````python
def div(a,b):
    try:
        return a/b
    except ZeroDivisionError as Error1:
        return '{}로 인한 계산식 표현 불가'.format(Error1)
    # except ZeroDivisionError as Error1:
    #     return '{}로 인한 계산식 표현 불가'.format(Error1)
div(1,0)
````

#enumerate

- 반복가능한 객체의 인덱스와 원소에 함께 접근할 수 있는 함수
- tuple(인덱스,원소)의 형태로 객체를 반환
- 보통 리스트 구조의 데이터일 때 활용한다.
- 반복문 사용시 몇 번째 반복문인지 확인이 필요할 때도 사용한다.


- list

````python
list1 = ['a', 'b', 'c']
x = enumerate(list1)
x
````

````python
list(x)
````

````python
next(x)
````

````python
list1 = ['a', 'b', 'c']
for x in enumerate(list1):
    print(x)
````

enumerate 활용예시  
enumerate는 for문에서 몇 번째 원소인지 확인할 수도 있지만,
enumerate 명령어를 활용하여 프로그래밍을 할 수도 있다.
- 예시: 병원에 방문한 순서대로 넘버링

enumerate를 사용하지 않은 일반 for문


````python
people = ['김씨', '최씨', '이씨']
i = 1
for j in people:
    print('번호: {}, 성: {}'.format(i, j))
    i = i + 1
````

enumerate를 사용한 for문

````python
people = ['김씨', '최씨', '이씨']
x = enumerate(people)
x
````

````python
next(x)
````

````python
list(x)
````

````python
people = ['김씨', '최씨', '이씨']
for x,y in enumerate(people):
    print(x+1,y)
````

````python
people = ['김씨', '최씨', '이씨']
for number, name in enumerate(people):
    print('번호: {}, 성: {}'.format(number+1, name))
````

#zip

- 동일한 개수로 이루어진 자료형을 묶어 주는 역할을 하는 함수
- 반복가능한 객체들을 병렬로 묶어주는 함수
- 각 원소들을 tuple구조로 묶음

````python
zip([1, 2, 3], [4, 5])
````

````python
list(zip([1, 2, 3], [4, 5])) # 매칭이 안되는 값 제거
````

````python
a = zip([1, 2, 3], [4, 5, 6])
````

````python
next(a)
````

````python
list(zip([1, 2, 3], [4, 5, 6], [7, 8, 9]))
````

````python
 list(zip(['one', 'two'],[1,2]))
````

````python
kor = ['사과', '바나나', '오렌지']
eng = ['apple', 'banana', 'orange']
for a,b in zip(kor, eng):
    # print(a)
    print('한글: {}, 영어: {}'.format(a,b))
````

#lambda

- lambda는 함수를 생성할 때 사용하는 예약어로 def와 동일한 역할을 한다.
- 기본구조 -> lambda 변수 : 수행할 문장


````python
def add(a,b):
    return a+b

add(1,2)
````

````python
add = lambda a, b: a+b

add(1, 2)
````

#map

- 반복문
- for문(반복문)과 map함수의 비교


예제) 실수가 저장된 리스트의 모든 요소를 문자로 변환


````python
a = [1.1, 2.3, 3.5, 4.1]
range(len(a)) # 0,1,2,3
````

````python
list(range(len(a)))
````

````python
a[0] = str(a[0])
a
````

````python
a = [1.1, 2.3, 3.5, 4.1]
for i in range(len(a)): #[0, 1, 2, 3]
    a[i] = str(a[i])
a
````

````python
a
````

- for 반복문으로 반복하면서 요소를 변환하는 것은 번거롭다.
- map 함수로 한줄로 간단하게 변환할 수 있다.


 map(함수, 자료)
 - map은 리스트(튜플, 문자 등)의 원소에 지정한 함수로 동일하게 처리해주는 함수

````python
a = [1.1, 2.3, 3.5, 4.1]
map(str,a)
list(map(str,a))
````

````python
a = [1.1, 2.3, 3.5, 4.1]
list(map(int,a))
````

- map과 lamda를 같이 쓸 수 있음
 - map(lamda 변수: 수행코드, 데이터)

````python
list(map(lambda a: '숫자'+str(a), [1.1, 2.6, 3.2, 4.8]))
````
